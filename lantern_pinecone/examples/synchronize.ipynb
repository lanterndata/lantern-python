```python
import lantern_pinecone
from getpass import getpass
```

```python
lantern_pinecone.init('postgres://postgres@localhost:5432/lantern')
```

```python
pinecone_ids = list(map(lambda x: str(x), range(1000000)))
index = lantern_pinecone.create_from_pinecone(
        api_key=getpass("Pinecone API Key"), 
        environment="us-east-1-aws",
        index_name="sift1m", 
        namespace="", 
        pinecone_ids=pinecone_ids, 
        recreate=True)
```

```python
index.describe_index_stats()
```

```python
index.query(top_k=10, include_values=True, include_metadata=True, id='455000', namespace="")
```
